```python
df = data.copy()
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoding = LabelEncoder()
df['type'] = encoding.fit_transform(df['type'].astype(str))
print (df.info())
```

```python
#the idea behind this is that all the transactions are distinct but they have M and C 
#at the starting that may have some importance in increasing the accuracy
def remove_numbers(transaction):
    for i in transaction:
        if i[0] == 'M':
            return 'M'
        if i[0] == 'C':
            return 'C'

# Apply the function to the columns
df['nameDest'] = df['nameDest'].apply(remove_numbers)
df['nameOrig'] = df['nameOrig'].apply(remove_numbers)


```

```python
df['nameDest'].value_counts()
```

```python
#to make the M and C in numeric form so that models can learn from them
df['nameDest'] = encoding.fit_transform(df['nameDest'].astype(str))
df['nameOrig'] = encoding.fit_transform(df['nameOrig'].astype(str))
print (df.info())
```

```python
df.head(10)
```

```python
# want to check whether there is any correlation between different columns or not, because if 
#there is it can decrease the accuracy of the model

from statsmodels.stats.outliers_influence import variance_inflation_factor

collinearity = pd.DataFrame()
collinearity["variables"] = df.columns
collinearity["Collinearity"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
print(collinearity)
```

```python
#making all the high collinear variables into one useful variable

df['Val_old_new_orig'] = df.apply(lambda func: func['oldbalanceOrg'] - func['newbalanceOrig'],axis=1)
df['Val_old_new_Dest'] = df.apply(lambda func: func['oldbalanceDest'] - func['newbalanceDest'],axis=1)

#Dropping columns
#nameOrig does have any correlation it has only one value that is Zero
df = df.drop(['oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest','nameOrig'],axis=1)

df.head()
```

```python
#after doing the processing again checking collinearity among the variables
collinearity = pd.DataFrame()
collinearity["variables"] = df.columns
collinearity["Collinearity"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
print(collinearity)
```

```python
#to check the correlation among the variables
corr=df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, cbar=True, square=True,annot=True, annot_kws={'size':8}, cmap='Reds')
```

```python
# for the model building I will be using the three prevalent models in this which are Logistic Regression
#, Random Forest and Decision Tree


from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import itertools
from collections import Counter
import sklearn.metrics as metrics
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
y = df["isFraud"]
x = df.drop(["isFraud"], axis= 1)

(x_train, x_test, y_train, y_test) = train_test_split(x, y, test_size= 0.2, random_state= 2)

```

```python
# Logistic Regression

lr = LogisticRegression()
lr.fit(x_train, y_train)

y_predict_lr = lr.predict(x_test)
lr_score = lr.score(x_test, y_test) * 100
```

```python
print(lr_score)
```

```python
# Evaluating the mode
# this is one of the most important thing of models 
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_predict_lr))

print("\nClassification Report:")
print(classification_report(y_test, y_predict_lr))
```

```python
# Random Forest Classifier

rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)

y_predict_rfc = rfc.predict(x_test)
rfc_score = rfc.score(x_test, y_test) * 100
```

```python
print(rfc_score)
```

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_predict_rfc))

print("\nClassification Report:")
print(classification_report(y_test, y_predict_rfc))
```

```python
# Decision Tree

dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)

y_predict_dtc = dtc.predict(x_test)
dtc_score = dtc.score(x_test, y_test) * 100
```

```python
print(dtc_score)
```

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_predict_dtc))

print("\nClassification Report:")
print(classification_report(y_test, y_predict_dtc))
```
